```python
import matplotlib.pyplot as plt
```

```python
x_sutun = ["Bacteria", "Virus", "Fungus", "Protozoa"]
y_sutun = [85, 86, 79, 80]
```

```python
max_y_lim = 100
min_y_lim = 0
plt.ylim(bottom = 10,top = 100)
plt.bar(x_sutun, y_sutun, color='rgmc')
plt.xticks(rotation=45)
plt.title(" Antigen Estimation Accuracy Rates for All Classes ")
plt.show
```
